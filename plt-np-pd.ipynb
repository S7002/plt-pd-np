```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas import DataFrame 
```

```python
employees=np.array([["Ahmed","28","System Admin",20000,3],["Fahad","24","Manager",40000,4.3],["Mohammed","26","IT support",10000,3.4],["Khalid","30","Web developer",15000,4]])
print(employees)
```

```python
df = pd.DataFrame(employees, columns=["Name", "Age", "Position", "Salary", "Rating"])
plt.figure(figsize=(5, 5))
plt.bar(df['Name'], df['Rating'].astype(float), color='gray')
plt.title('Employee of the month')
plt.xlabel('Employee Name')
plt.ylabel('Rate ★')
plt.ylim(0,5)
plt.show()
```
